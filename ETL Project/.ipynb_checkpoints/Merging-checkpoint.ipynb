```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Name of the CSV file
file_cad = 'can_cleaned_final.csv'
file_usa = 'USA_cleaned_final.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
df_cad = pd.read_csv(file_cad, encoding="ISO-8859-1")
df_usa = pd.read_csv(file_usa, encoding="ISO-8859-1")
```

```python
# Preview of the DataFrame
# Note that FIELD8 is likely a meaningless column
df_cad.head()
```

```python
#Preview of the DataFrame
# Note that FIELD8 is likely a meaningless column

df_usa.head()
```

```python
# Merge two dataframes using an inner join
merge_table = pd.merge(df_usa, df_cad, on=["Origin","Year"])
merge_table.head()
```

```python
# Organize the columns so they are in a more logical order
merge_table_org = merge_table[[
    "Year", "Origin", "Total Population_x", "Total Population_y"]]

merge_table_org.head()
```

```python
# Rename columns for readability
merge_table_renamed = merge_table_org.rename(columns={"Total Population_x": "Population in USA",
                                                "Total Population_y": "Population in CAD",
                                                                                                })

merge_table_renamed.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
merge_table_renamed.to_csv("Merged_file_final.csv", index=False, header=True)
```
